```python
#Day 1 Puzzle 1
f = open(r'C:\Users\19177\Desktop\Python Projects\Programming-Puzzles\Advent of Code\Advent of Code Input\Advent Day 1.txt','r')
lines = f.readlines()

#Convert strings to integers
list_1 = len(lines)
lines_int = lines
for i in range(0, list_1):
    lines_int[i] = int(lines[i])

#keep a count
count_plus = 0

#Count the number of ascensions 
for i in range(1, list_1):
    if lines_int[i]-lines_int[i-1]>0:
        count_plus +=1
# D1P1 Solution        
print('D1P1 Solution:' + str(count_plus))


#D1P2
#Count ascensions in a sliding group of 3
lines_slide = []

for i in range(2,list_1):
    lines_slide.append(lines_int[i] + lines_int[i-1] + lines_int[i-2])

#keep a count
count_plus_slide = 0

#Count the number of ascensions 
for i in range(1, len(lines_slide)):
    if lines_slide[i]-lines_slide[i-1]>0:
        count_plus_slide +=1
#D1P2 Solution
print('D1P2 Solution:' + str(count_plus_slide))
```

```python
#D2P1
f = open(r'C:\Users\19177\Desktop\Python Projects\Programming-Puzzles\Advent of Code\Advent of Code Input\Advent Day 2.txt','r')
lines = f.readlines()

#record useful info
directions = [(lines[i][0],int(lines[i][-2])) for i in range(0,len(lines)-1)]
#last entry has different format from previous entries
directions.append((lines[len(lines)-1][0],int(lines[len(lines)-1][-1])))

Hp = 0
dp = 0

for i in range(0,len(directions)):
    if directions[i][0] == 'f':
        Hp += directions[i][1]
    elif directions[i][0] == 'u':
        dp -= directions[i][1]
    else:
        dp += directions[i][1]
        
print('D2P1 Solution:' + str(Hp*dp))

#D2P2
Hp = 0
dp = 0
aim = 0

for i in range(0,len(directions)):
    if directions[i][0] == 'f':
        Hp += directions[i][1]
        dp += directions[i][1]*aim
    elif directions[i][0] == 'u':
        aim -= directions[i][1]
    else:
        aim += directions[i][1]
        
print('D2P1 Solution:' + str(Hp*dp))
```

```python
#Day 3

def Parse(word):
    return [int(char) for char in word.rstrip()]

def power(list_bits):
    gamma = [int(round(idx/len(list_bits))) for idx in np.sum(list_bits,0)]
    gamma_int = int("".join([str(char) for char in gamma]),2)
    epsilon_int = int("".join([str(1-char) for char in gamma]),2)
    return gamma, epsilon, gamma_int*epsilon_int

def life_support_rating(list_bits):
    
    oxy = list_bits
    for idx_2 in range(0, len(oxy[0])):
        oxy = [idx for idx in oxy if idx[idx_2]==int(np.sum(oxy,0)[idx_2]/len(oxy)+.5)]
        if len(oxy)==1:
            break
    co2 = list_bits
    for idx_2 in range(0, len(co2[0])):
        co2 = [idx for idx in co2 if idx[idx_2]==1-int(np.sum(co2,0)[idx_2]/len(co2)+.5)]
        if len(co2)==1:
            break
    #Avg() would have worked above too.
    #round() rounds to nearest even integer when arg is x.5
    
    oxy_int = int("".join([str(char) for char in oxy[0]]),2)
    co2_int = int("".join([str(char) for char in co2[0]]),2)
    
    return oxy, co2, oxy_int*co2_int

f = open(r'C:\Users\19177\Desktop\Python Projects\Programming-Puzzles\Advent of Code\Advent of Code Input\Advent Day 3.txt','r')
input = [Parse(line) for line in f.readlines()]

(gamma, epsilon, power) = power(input)
(oxy, co2, lsr) = life_support_rating(input)

print('D3P1 Solution:' + str(power))
print('D3P2 Solution:' + str(lsr))
```

```python
#Day 4
def bingo_board(values):
    #import a list of 5 strings each of 5 values
    bingo_card = []
    #Need the strip to get rid of white spaces and 'if s' gets rid of empty strings.
    for row in range(len(values)):
        bingo_row = [int(idx.strip()) for idx in values[row].split(" ") if idx]
        bingo_card.append(bingo_row)
    #Add the columns as rows to make them easier to deal with
    [bingo_card.append(col) for col in np.array(bingo_card).T.tolist()]    
    #sorted rows should reduce time complexity later
    [bingo_card[idx].sort() for idx in range(len(bingo_card))]
    return bingo_card

def cross_numbers(boards, number, boards_done):
    boards_done_now=[]
    board_scores_new=np.zeros(len(boards_done))
    done = False
    winning_card = 0
    for idx in range(len(boards)):
        #skip winning boards
        if boards_done[int((idx)/10)] == 1:
            continue
        
        for idx_2 in range(len(boards[idx])):
            if type(boards[idx][idx_2]) == str:
                continue
            elif boards[idx][idx_2] == number:
                boards[idx][idx_2] = str(boards[idx][idx_2])
                if check_winner(boards[idx]):
                    winning_card = int((idx+1)/10)+1
                    print("Board " + str(winning_card) + " has finished with a score of " 
                          + str(score(boards[10*int((idx+1)/10):10*int((idx+1)/10)+5],  number)))
                    done = True
                    board_scores_new[winning_card-1] = score(boards[10*int((idx+1)/10):10*int((idx+1)/10)+5],  number)
                    boards_done_now.append(int((idx+1)/10))
                break
            elif boards[idx][idx_2] > number:
                #Rows are sorted, so this skips unnecessary checks
                break
            else:
                continue
    #print(len(boards_done_now))
    return boards, done, winning_card, boards_done_now, board_scores_new

def check_winner(row):
    return all(isinstance(idx,str) for idx in row)

def score(board, number):
    row_sums = []
    
    for idx_2 in range(len(board)):
        row_sums.append(sum([idx for idx in board[idx_2] if not isinstance(idx,str)]))
                        
    return sum(row_sums)*number

f = open(r'C:\Users\19177\Desktop\Python Projects\Programming-Puzzles\Advent of Code\Advent of Code Input\Advent Day 4.txt','r')
input = [line for line in f.readlines()]

#get the bingo numbers
numbers = [int(idx) for idx in input[0].split((","))]

#get the bingo cards; new card every 10 rows
bingo_cards = []
for idx in range(2, len(input), 6):
    [bingo_cards.append(row) for row in bingo_board(input[idx:idx+5])]

boards_done = np.zeros(int(len(bingo_cards)/10))
board_scores = np.zeros(int(len(bingo_cards)/10))
for idx in range(len(numbers)):
    bingo_cards, done, winning_card, boards_done_now, board_scores_new = cross_numbers(bingo_cards, numbers[idx], boards_done)
    boards_done[boards_done_now] = 1
    board_scores[boards_done_now] = board_scores_new[boards_done_now]
    #Stop when all boards are done
    if len(boards_done) == sum(boards_done):
        #print(idx)
        break
        
        
#print('D4P1 Solution:' + str(scores[0]))
#print('D4P2 Solution:' + str(scores[-1]))
```
