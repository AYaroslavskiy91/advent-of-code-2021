```python
import numpy as np
from collections import deque

def value(char):
    value = {')':3,
            ']':57,
            '}': 1197,
            '>':25137}
    return value[char]

def value_p2(char):
    value = {'(':1,
            '[':2,
            '{':3,
            '<':4}
    return value[char]

def parse_line(bracket_line):
    openers = ['[', '{', '(', '<']
    closers = [']', '}', ')', '>']
    pairs = [(openers[idx], closers[idx]) for idx in range(len(openers))]
    
    bracket_stack = deque()
    for char in bracket_line:
            if char in openers:
                bracket_stack.append(char)
            else:
                beg = bracket_stack.pop()
                if (beg, char) not in pairs:
                    print("error!" + char)
                    return (value(char), 0)
    print("Incomplete")
    value2 = 0
    bracket_stack.reverse()
    for idx in range(len(bracket_stack)):
        value2 = 5*value2 + value_p2(bracket_stack[idx])
    return (0, value2)
    
f = open(r'C:\Users\19177\Desktop\Python Projects\Programming-Puzzles\Advent of Code\Advent of Code Input\Advent Day 10.txt','r')
lines = f.readlines()

part_1_score = 0
part_2_scores = []
for line in lines:
    bracket_line = line.split()[0]
    p1, p2 = parse_line(bracket_line)
    part_1_score += p1
    part_2_scores.append(p2)
print(part_1_score)
part_2_scores = [idx for idx in part_2_scores if idx!=0]
part_2_scores.sort()
print(part_2_scores[int((len(part_2_scores)-1)/2)])
```
